# <b> Classificação Automática de Imagens de Células Sanguíneas usando Redes Neurais Convolucionais </b>

###### <b>Descrição do conjunto de dados:</b> 12.500 imagens de células sanguíneas (JPEG) acompanhadas de rótulos de tipo de célula (CSV). Este conjunto de dados é composto por cerca de 3.000 imagens para cada um dos 4 tipos de células diferentes agrupadas em 4 pastas diferentes (de acordo com o tipo de célula). Os tipos de células são:

<b>1. Eosinófilos<p></b>
<b>2. Linfócitos<p></b>
<b>3. Monócitos<p></b>
<b>4. Neutrófilos<p></b>

### <b> Importando bibliotecas e configurando ambiente </b>

```python
# importando tensorflow
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPool2D
from tensorflow.keras.layers import ZeroPadding2D
from tensorflow.keras.layers import Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam

# bibliotecas para manipulação de arrays
import numpy as np

# bibliotecas para plotagem
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# bibliotecas para manipulação de imagens
import cv2

# bibliotecas para manipulação de diretórios
import os
```

```python
# desativando o ambiente CUDA
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"
```

### <b> Importando o conjunto de dados </b>

```python
# Carregando as imagens
imagem1 = mpimg.imread('blood_cell_image_dataset/dataset2-master/dataset2-master/images/TRAIN/EOSINOPHIL/_0_1169.jpeg')
imagem2 = mpimg.imread('blood_cell_image_dataset/dataset2-master/dataset2-master/images/TRAIN/LYMPHOCYTE/_0_5620.jpeg')
imagem3 = mpimg.imread('blood_cell_image_dataset/dataset2-master/dataset2-master/images/TRAIN/MONOCYTE/_0_2232.jpeg')
imagem4 = mpimg.imread('blood_cell_image_dataset/dataset2-master/dataset2-master/images/TRAIN/NEUTROPHIL/_1_3725.jpeg')

# criando subgráficos
fig, axs = plt.subplots(1, 4, figsize=(12, 3))

# Desativando os eixos para cada subgráfico
for ax in axs:
    ax.axis('off')

# Adicionando cada imagem aos subgráficos
axs[0].imshow(imagem1)
axs[0].set_title('EOSINOPHIL')

axs[1].imshow(imagem2)
axs[1].set_title('LYMPHOCYTE')

axs[2].imshow(imagem3)
axs[2].set_title('MONOCYTE')

axs[3].imshow(imagem4)
axs[3].set_title('NEUTROPHIL')

# Ajustando o layout para evitar sobreposição
plt.tight_layout()

# Exibindo a figura
plt.show()
```

```python
# Caminho para a pasta principal
conjunto_de_dados = 'blood_cell_image_dataset/dataset2-master/dataset2-master/images/TRAIN'

# Lista para armazenar a contagem de elementos em cada classe
contagens_treinamento = []

# Lista de nomes das subpastas
subpastas = os.listdir(conjunto_de_dados)

# Definir o tamanho da figura (largura x altura)
fig, ax = plt.subplots(figsize=(6, 6/1.618))

# Contagem de imagens em cada pasta
for subpasta in subpastas:
    caminho_subpasta = os.path.join(conjunto_de_dados, subpasta)
    
    # Contagem de elementos na subpasta
    contagem_elementos = len(os.listdir(caminho_subpasta))
    
    # Adicionar contagem à lista
    contagens_treinamento.append(contagem_elementos)

# Criar um histograma
plt.bar(subpastas, contagens_treinamento, color='blue')
plt.xlabel('Classes')
plt.ylabel('Quantidade de imagens')
plt.ylim(2350, 2550)
plt.title('Conjunto de dados (Treinamento)')

# Adicionando rótulos de dados ao topo de cada barra
for i, valor in enumerate(contagens_treinamento):
    plt.text(i, valor + 5, str(valor), ha='center', va='bottom')
    
plt.show()
```

```python
# Caminho para a pasta principal
conjunto_de_dados = 'blood_cell_image_dataset/dataset2-master/dataset2-master/images/TEST'

# Lista para armazenar a contagem de elementos em cada classe
contagens_teste = []

# Lista de nomes das subpastas
subpastas = os.listdir(conjunto_de_dados)

# Definir o tamanho da figura (largura x altura)
fig, ax = plt.subplots(figsize=(6, 6/1.618))

# Contagem de imagens em cada pasta
for subpasta in subpastas:
    caminho_subpasta = os.path.join(conjunto_de_dados, subpasta)
    
    # Contagem de elementos na subpasta
    contagem_elementos = len(os.listdir(caminho_subpasta))
    
    # Adicionar contagem à lista
    contagens_teste.append(contagem_elementos)

# Criar um histograma
plt.bar(subpastas, contagens_teste, color='blue')
plt.xlabel('Classes')
plt.ylabel('Quantidade de imagens')
plt.ylim(550, 660)
plt.title('Conjunto de dados (Teste)')

# Adicionando rótulos de dados ao topo de cada barra
for i, valor in enumerate(contagens_teste):
    plt.text(i, valor + 5, str(valor), ha='center', va='bottom')
    
plt.show()
```

```python
# preparando os dados
img_width, img_height = 64, 64
datagen = ImageDataGenerator(rescale = 1/255.0, validation_split = 0.25)
train_data_generator = datagen.flow_from_directory(directory="blood_cell_image_dataset/dataset2-master/dataset2-master/images/TRAIN",
                                                   target_size = (img_width, img_height), color_mode="rgb",
                                                   class_mode="categorical", batch_size = 16, shuffle=True ,subset = "training")
validation_data_generator = datagen.flow_from_directory(directory="blood_cell_image_dataset/dataset2-master/dataset2-master/images/TRAIN",
                                                        target_size = (img_width, img_height),  color_mode="rgb",
                                                        class_mode="categorical", batch_size = 16, shuffle=True, subset = "validation")
```

```python
quantidade_conjunto_de_dados = sum(contagens_treinamento) + sum(contagens_teste)
quantidade_conjunto_de_dados_treinamento = len(train_data_generator.filenames)
quantidade_conjunto_de_dados_validacao = len(validation_data_generator.filenames)
quantidade_conjunto_de_dados_teste = sum(contagens_teste)

print(f'Conjunto de dados completo: {quantidade_conjunto_de_dados} imagens')
print(f'Conjunto de dados de treinamento: {quantidade_conjunto_de_dados_treinamento}   imagens --- {quantidade_conjunto_de_dados_treinamento/quantidade_conjunto_de_dados * 100:.2f} %')
print(f'Conjunto de dados de validação:   {quantidade_conjunto_de_dados_validacao}   imagens --- {quantidade_conjunto_de_dados_validacao/quantidade_conjunto_de_dados * 100:.2f} %')
print(f'Conjunto de dados de teste:       {quantidade_conjunto_de_dados_teste}   imagens --- {quantidade_conjunto_de_dados_teste/quantidade_conjunto_de_dados * 100:.2f} %')
```

### <b> CNN com função de atição ReLu </b>

```python
# criando o modelo de classificação CNN
model_relu = Sequential()

model_relu.add(Conv2D(16, (3,3), input_shape=(img_width,img_height,3), activation="relu", padding='valid'))
model_relu.add(MaxPool2D(2,2))

model_relu.add(Conv2D(32, (3,3), activation="relu", padding='valid'))
model_relu.add(MaxPool2D(2,2))

model_relu.add(Conv2D(64, (3,3), activation="relu", padding='valid'))
model_relu.add(MaxPool2D(2,2))

model_relu.add(Conv2D(128, (3,3), activation="relu", padding='valid'))
model_relu.add(MaxPool2D(2,2))

model_relu.add(Flatten())

model_relu.add(Dense(128, activation="relu"))

model_relu.add(Dense(4, activation="softmax"))
```

```python
model_relu.summary()
```

```python
model_relu.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history_relu = model_relu.fit(train_data_generator, steps_per_epoch=len(train_data_generator),
                         epochs=50, validation_data=validation_data_generator,
                         validation_steps = len(validation_data_generator))
```

```python
# gráfico de acurácia
fig, ax = plt.subplots(figsize=(8, 6/1.618))
plt.plot(history_relu.history['accuracy'])
plt.plot(history_relu.history['val_accuracy'])
plt.title('Função de Ativação ReLu')
plt.ylabel('Acurácia')
plt.xlabel('Época')
plt.legend(['Treinamento', 'Validação'], loc='lower right')
plt.show()
```

```python
# gráfico de perda
fig, ax = plt.subplots(figsize=(8, 6/1.618))
plt.plot(history_relu.history['loss'])
plt.plot(history_relu.history['val_loss'])
plt.title('Função de ativação ReLu')
plt.ylabel('Função de perda')
plt.xlabel('Época')
plt.legend(['Treinamento', 'Validação'], loc='upper right')
plt.show()
```

```python
# calculando a acurácia no teste
datagen_test = ImageDataGenerator(rescale = 1/255.0)
test_data_generator = datagen.flow_from_directory(directory="blood_cell_image_dataset/dataset2-master/dataset2-master/images/TEST/",
                                                  target_size = (img_width, img_height), color_mode="rgb",
                                                  class_mode="categorical", batch_size = 16, subset = "training")
model_relu.evaluate(test_data_generator)
```
